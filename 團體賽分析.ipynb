```python
import itertools as it

N = 8 # must be even
all_places = [i+1 for i in range(N)]
all_possibles = it.combinations(all_places, int(N/2))
weight = dict(zip(all_places, [10,8,6,5,4,3,2,1]))
def score(places, weight) -> int:
  sum = 0
  for p in places:
    sum += weight[p]
  return sum  
s = 0
win__possibles = set()
for A in all_possibles:
  s += 1
  B = [i for i in all_places if i not in A]
  print(f"A: {A} B: {tuple(B)}")
  print(f"A score: {score(A, weight)}\nB score: {score(tuple(B), weight)}")
  print("A win\n" if score(A, weight)>score(tuple(B), weight) else "B win\n")
  if score(A, weight)>score(tuple(B), weight):
    win__possibles.add(A)
  if score(A, weight)<score(tuple(B), weight):
    win__possibles.add(tuple(B))
  if score(A, weight)==score(tuple(B), weight):  
    raise "could be draw"
print(f"Total possible count: {s}")
```
